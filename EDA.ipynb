```python
import nibabel as nib

# Load the .hdr/.img file (use .hdr file, it automatically loads .img)
file_path = "/Users/poulam/Desktop/Brain Age prediction/disc1/OAS1_0029_MR1/processed/MPRAGE/T88_111/OAS1_0019_MR1_mpr_n4_anon_111_t88_masked_gfc.hdr"  # Replace with your actual file path
img = nib.load(file_path)

# Print general information
print(img)  # Shows metadata, shape, data type

# Get data array (3D NumPy array)
data = img.get_fdata()

# Print shape of the image
print("Image shape:", data.shape)

# Print voxel size (spatial resolution)
print("Voxel size:", img.header.get_zooms())

# Print header information (metadata)
print("Header info:\n", img.header)
```

```python
import matplotlib.pyplot as plt

# Get voxel data
data = img.get_fdata()
n=73
# Select a central slice along the Z-axis
z_slice = data[:, n, :]  # nth slice (middle)

# Display the image
plt.imshow(z_slice, cmap="gray")
plt.colorbar()
plt.title(f"MRI Slice at Z={n}")
plt.show()
```
